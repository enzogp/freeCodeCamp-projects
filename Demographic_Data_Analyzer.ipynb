# What is the average age of men?

```python
df[df["sex"] == ' Male']["age"].mean()
```

# What is the percentage of people who have a Bachelor's degree?

```python
p_bachelor = (df['education'] == ' Bachelors').sum() / len(df) * 100
print(p_bachelor)
```

# What percentage of people with advanced education (Bachelors, Masters, or Doctorate) make more than 50K?

```python
higher_education = df['education'].isin([' Bachelors', ' Masters', ' Doctorate'])
rich = df['salary'] == ' >50K'

advanced_edu_rich = (df[higher_education & rich].shape[0] / df[higher_education].shape[0]) * 100
print(advanced_edu_rich)
```

# What is the minimum number of hours a person works per week?

# What percentage of the people who work the minimum number of hours per week have a salary of more than 50K?

```python
min_hours = df["hours_per_week"] == 1
rich = df["salary"] == " >50K"

f = (df[min_hours & rich].shape[0] / df[min_hours].shape[0]) *100
print(f)
```

# What country has the highest percentage of people that earn >50K and what is that percentage?

```python
rich = df['salary'] == ' >50K'

total_per_country = df["native_country"].value_counts()

rich_per_country = df[rich]["native_country"].value_counts()

p_rich_per_country = (rich_per_country / total_per_country) *100

top_country = p_rich_per_country .idxmax()
top_percent = round(p_rich_per_country .max(), 1)

print(top_country)
print(top_percent)
```
